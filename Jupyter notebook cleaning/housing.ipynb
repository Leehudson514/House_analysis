```python
# upload metro housing data
full_house = "../Resources/ALL_zipcode_rent.csv"
full_house_df = pd.read_csv(full_house)
full_house_df.head()
```

```python
#change 'date' str to date
full_house_df['Date'] = pd.to_datetime(full_house_df['Date'])
full_house_df.head()
```

```python
#format currency and import babel
from babel.numbers import format_currency

full_house_df["Rent"] = full_house_df["Rent"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
full_house_df.head()
```

```python
#update column names
full_house_df = full_house_df.rename(columns={'MsaName':'city'})
full_house_df.head()
```

```python
#save csv
full_house_df.to_csv('usa_house_rent.csv', index = False)
```

```python
# get year from the corresponding
# date column value
full_house_df['year'] = pd.DatetimeIndex(full_house_df['Date']).year
 
# get month from the corresponding
# ate column value
full_house_df['month'] = pd.DatetimeIndex(full_house_df['Date']).month

full_house_df.head()
```

```python
full_house_df.to_csv('usa_house_rent.csv', index = False)
```
