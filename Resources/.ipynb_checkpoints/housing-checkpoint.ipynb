```python
# Import
import pandas as pd
```

```python
# upload metro housing data
data_file = "Metro_ZORI_AllHomesPlusMultifamily_Smoothed.csv"
data_file_pd = pd.read_csv(data_file)
data_file_pd.head()
```

```python
#upload zip code housing data
zip_data = "Zip_ZORI_AllHomesPlusMultifamily_Smoothed.csv"
zip_data_df = pd.read_csv(zip_data)
zip_data_df.head()
```

```python
# count the rows and columns in a DataFrame
zip_data_df = zip_data_df.dropna()
print("\nDataFrame after removing null values...\n",zip_data_df)
print("\n(Updated) Number of rows and column in our DataFrame = ",zip_data_df.shape)
```

```python
# count the rows and columns in a DataFrame
data_file_pd = data_file_pd.dropna()
print("\nDataFrame after removing null values...\n",data_file_pd)
print("\n(Updated) Number of rows and column in our DataFrame = ",data_file_pd.shape)
```

```python
#convert wide to long
zip_df = zip_data_df.melt(id_vars=['RegionID','RegionName','SizeRank','MsaName'], var_name='Date',value_name='Rent')
```

```python
#display df
zip_df.head()
```

```python
# convert wide to long
metro_df = data_file_pd.melt(id_vars=['RegionID','RegionName','SizeRank'], var_name='Date',value_name='Rent')
metro_df.head()
```

```python
#save to csv
zip_df.to_csv('Clean_zipcode_rent.csv', index = False)
```

```python
# upload metro housing data
full_house= "Full_house.csv"
full_house_df = pd.read_csv(full_house)
full_house_df.head()
```

```python
#change 'date' str to date
full_house_df['date'] = pd.to_datetime(full_house_df['date'])
full_house_df.head()
```

```python
#format currency and import babel
from babel.numbers import format_currency

full_house_df["rentavg"] = full_house_df["rentavg"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
full_house_df.head()
```

```python
#update column names
full_house_df = full_house_df.rename(columns={'zip':'zipCode','msaname':'city','rentavg':'rent'})
full_house_df.head()
```

```python
#save csv
full_house_df.to_csv('usa_house_rent.csv', index = False)
```

```python
# get year from the corresponding
# date column value
full_house_df['year'] = pd.DatetimeIndex(full_house_df['Date']).year
 
# get month from the corresponding
# ate column value
full_house_df['month'] = pd.DatetimeIndex(full_house_df['Date']).month

full_house_df.head()
```
